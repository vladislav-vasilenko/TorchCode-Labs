# 🟠 Medium: Cosine LR Scheduler with Warmup

Реализуйте **cosine learning-rate schedule с linear warmup**. Scheduler не обновляет параметры сам: он только возвращает learning rate для текущего номера шага. Warmup делает первые обновления осторожными, затем cosine decay плавно уменьшает шаг до `min_lr`.

### Две фазы
1. **Warmup:** от `0` на `step=0` до `max_lr` на границе `step=warmup_steps`.
2. **Cosine decay:** progress меняется от `0` до `1`, а learning rate — от `max_lr` до `min_lr`.

Функция должна быть непрерывной на границе: обе ветви при `step=warmup_steps` дают `max_lr`. В конце `cos(π)=-1`, поэтому при `step=total_steps` получается `min_lr`.


### Формулы schedule
```
step < warmup:  lr = max_lr * step / warmup_steps  (linear ramp)
step >= warmup: lr = min_lr + 0.5*(max_lr-min_lr)*(1 + cos(π * progress))
```
где `progress = (step - warmup_steps) / (total_steps - warmup_steps)`.

### План реализации
1. Выберите ветвь по сравнению `step` и `warmup_steps`.
2. В warmup вычислите линейную долю пути до `max_lr`.
3. После warmup нормализуйте step в progress от `0` до `1`.
4. Примените cosine к диапазону `max_lr - min_lr`, затем добавьте `min_lr`.

Продумайте edge cases `warmup_steps=0` и `total_steps=warmup_steps`, чтобы не получить деление на ноль; основной judge проверяет обычный двухфазный случай.

### Быстрые самопроверки
- `step=0 → 0`;
- `step=warmup_steps → max_lr`;
- середина cosine-фазы даёт середину между `max_lr` и `min_lr`;
- `step=total_steps → min_lr`;
- warmup монотонно растёт, decay монотонно убывает.

### Типичные ошибки
- off-by-one на границе warmup;
- progress делится на `total_steps`, а не на длину decay-фазы;
- забытый множитель `π`;
- cosine decay начинается с `min_lr`;
- integer division в языках, где `/` не даёт float.
