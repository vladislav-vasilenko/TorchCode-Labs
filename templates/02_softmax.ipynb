# 🟢 Easy: Implement Softmax

Реализуйте **Softmax** с помощью базовых операций. Softmax преобразует произвольные logits в распределение: значения становятся неотрицательными, а сумма вдоль выбранной оси равна единице.

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}$$

### Что означает `dim`
Для входа формы `(batch, classes)` обычно используют `dim=-1`: каждая строка получает своё распределение по классам. Для тензора любой размерности нормализация выполняется независимо для каждого среза вдоль `dim`.

### Численная устойчивость
Прямой `exp(x)` может переполниться уже на больших положительных logits. Softmax не меняется, если вычесть из всех элементов среза одну константу:

$$\text{softmax}(x) = \text{softmax}(x - \max(x))$$

Максимум нужно вычислять вдоль того же `dim` и сохранять размерность (`keepdim=True`), иначе broadcasting может работать неверно.

### План реализации
1. Найдите максимум каждого среза вдоль `dim`.
2. Вычтите максимум, затем вычислите экспоненты.
3. Просуммируйте экспоненты вдоль `dim` с сохранением размерности.
4. Разделите каждый элемент на соответствующую сумму.

### Ограничения
- не используйте `torch.softmax`, `F.softmax` или `torch.nn.Softmax`;
- реализация должна быть численно устойчивой и сохранять форму входа.

### Быстрые самопроверки
- сумма результата вдоль `dim` близка к `1`;
- все значения лежат между `0` и `1`;
- `my_softmax(x)` совпадает с `my_softmax(x + 1000)` с разумной точностью;
- для одинаковых logits получается равномерное распределение.

```
Input:  tensor([1., 2., 3.])
Output: tensor([0.0900, 0.2447, 0.6652])  # sums to 1.0
```

### Типичные ошибки
- редукция по неверной оси;
- потеря размерности после `max` или `sum`;
- экспонента до вычитания максимума;
- одна общая сумма для всего batch вместо отдельной суммы на объект.
