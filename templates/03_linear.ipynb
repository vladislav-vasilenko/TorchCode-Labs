# 🟡 Medium: Simple Linear Layer

Реализуйте простой fully-connected слой: **`y = xWᵀ + b`**. Каждый output feature — взвешенная сумма всех input features плюс собственный bias. Это базовый строительный блок MLP, projections attention и classifier heads.

### Ориентация weight
`weight` хранится как `(out_features, in_features)`: каждая строка описывает один output neuron. Вход `x` заканчивается dimension `in_features`, поэтому для матричного умножения используется `weight.T` формы `(in_features, out_features)`.

Для `x` формы `(B, in_features)` результат матричного умножения имеет `(B, out_features)`. Bias формы `(out_features,)` broadcast по batch. Та же логика работает для дополнительных leading dimensions, например `(B,T,in_features)`.

### Почему важна инициализация
Случайные значения стандартного масштаба слишком велики при большом `in_features`: сумма многих слагаемых увеличивает variance. Масштаб `1/sqrt(in_features)` удерживает output activations в разумном диапазоне.

### Контракт
```python
class SimpleLinear:
    def __init__(self, in_features: int, out_features: int): ...
    def forward(self, x: torch.Tensor) -> torch.Tensor: ...
```

### Требования
- `self.weight`: `(out_features, in_features)`, initialization `randn * (1/√in_features)`;
- `self.bias`: `(out_features,)`, initialization zeros;
- оба tensor должны иметь `requires_grad=True`;
- `forward(x)` вычисляет `x @ W.T + b`;
- не используйте `torch.nn.Linear`.

### План реализации
1. Создайте weight с нужной ориентацией и масштабом.
2. Создайте нулевой bias.
3. В forward выполните одно матричное умножение и broadcasting bias.
4. Не копируйте параметры через операции, разрывающие autograd.

### Быстрые самопроверки
- shapes параметров совпадают с контрактом;
- выход `(2,8) → (2,4)` при `in=8`, `out=4`;
- вручную вычисленный `x @ weight.T + bias` совпадает с forward;
- после backward у weight и bias есть gradients;
- при нулевом input output равен bias.

### Типичные ошибки
- weight формы `(in,out)` вместо `(out,in)`;
- забытый transpose;
- bias формы `(1,out)` без необходимости;
- parameters создаются с `requires_grad=False`;
- слишком крупная initialization без деления на `sqrt(in_features)`.
