# 🟢 Easy: Implement ReLU

Реализуйте **ReLU** (Rectified Linear Unit) с помощью базовых тензорных операций. Это первая хорошая практика поэлементной функции: форма тензора не меняется, а правило применяется независимо к каждому элементу.

$$\text{ReLU}(x) = \max(0, x)$$

### Что важно понять
- для отрицательных значений выход равен нулю, для положительных — самому входу;
- производная равна `0` слева от нуля и `1` справа, поэтому операции должны оставаться в графе autograd;
- в точке `x = 0` производная математически неоднозначна; PyTorch использует согласованное практическое значение;
- вход и выход имеют одинаковые `shape`, `dtype` и `device`.

### План реализации
1. Постройте булеву маску положительных элементов.
2. По маске выберите исходное значение или ноль совместимого типа.
3. Не преобразуйте тензор в Python-список и не создавайте результат через `torch.tensor(...)`: это разорвёт граф вычислений.

### Ограничения
- не используйте `torch.relu`, `F.relu`, `torch.clamp` и готовые activation-модули;
- результат должен поддерживать autograd.

### Быстрые самопроверки
- результат для отрицательного вектора состоит из нулей;
- положительные элементы не изменяются;
- форма многомерного входа сохраняется;
- после `relu(x).sum().backward()` у leaf-тензора `x` заполнен `x.grad`.

```
Input:  tensor([-2., -1., 0., 1., 2.])
Output: tensor([ 0.,  0., 0., 1., 2.])
```

### Типичные ошибки
- изменение входа in-place без необходимости;
- цикл по отдельным значениям вместо векторной операции;
- создание CPU-нуля, несовместимого с `dtype` или `device` входа.
