# Треки TorchCode Labs

Это оглавление объединяет задачи апстрима и новые учебные треки форка. Проходите ноутбуки внутри выбранного трека в указанном порядке.

| Трек | Цель | Порядок ноутбуков | Статус |
|---|---|---|---|
| `core` | Отработать базовые операции PyTorch, attention, архитектуры, обучение и декодирование. | `01` → `40` | ✅ Реализован: 40 задач и теория (spec 040) |
| `distributed` | Разобраться в DDP, collective operations, памяти оптимизаторов, ZeRO/FSDP и честном измерении шага. | `5X` | ⬜ Запланирован: specs 101–109 |
| `inference` | Освоить практики эффективного production-инференса и оптимизации декодирования. | `6X` | 📝 Спеки ещё не написаны |
| `nanogpt` | Собрать ключевые компоненты компактного GPT и связать их в обучающий пайплайн. | `7X` | 📝 Спеки ещё не написаны |
| `positional` | Сравнить позиционные представления и понять их влияние на attention и длину контекста. | `7X` | 📝 Спеки ещё не написаны |

Задачи без явно заданного поля `track` относятся к `core`. Для ещё не реализованных треков диапазон обозначает зарезервированную нумерацию, а не существующие файлы.
