# 🟢 Easy: Kaiming Initialization

Реализуйте **Kaiming (He) normal initialization** для весов слоя перед ReLU. Инициализация задаёт масштаб случайных весов так, чтобы дисперсия активаций не затухала и не взрывалась при прохождении через много слоёв.

$$W \sim \mathcal{N}(0, \text{std}^2) \quad \text{where} \quad \text{std} = \sqrt{\frac{2}{\text{fan\_in}}}$$

### Откуда берётся коэффициент 2
`fan_in` — число входов одного нейрона. Для линейного слоя PyTorch хранит weight в форме `(fan_out, fan_in)`. ReLU зануляет значительную часть симметрично распределённых входов, поэтому gain `sqrt(2)` компенсирует уменьшение дисперсии.

В этой задаче используется режим `fan_in`, который ориентирован на сохранение масштаба forward-активаций, и нормальное распределение со средним ноль.

- shape и identity объекта после вызова не изменяются;
- на большой матрице среднее близко к `0`;
- выборочное std близко к `sqrt(2/fan_in)`;

### Типичные ошибки
- использование `weight.shape[0]` вместо `shape[1]`;
- путаница между variance `2/fan_in` и standard deviation `sqrt(2/fan_in)`;
- возврат нового тензора вместо изменения переданного;
- случайная запись операции инициализации в autograd-граф. В реальных модулях инициализацию выполняют без отслеживания градиентов.
