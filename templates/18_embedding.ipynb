# 🟢 Easy: Embedding Layer

Реализуйте **Embedding Layer** как обучаемую таблицу векторов. Каждый целочисленный индекс выбирает одну строку матрицы весов. Это не умножение на число: индекс используется как адрес строки.

### Представление
Пусть словарь содержит `V = num_embeddings` объектов, а размер вектора равен `D = embedding_dim`. Тогда таблица имеет форму `(V, D)`. Для индексов формы `(B, T)` результат имеет форму `(B, T, D)`: к форме индексов добавляется последняя embedding-размерность.

Выбор строки эквивалентен умножению one-hot вектора длины `V` на матрицу весов, но прямой lookup не создаёт огромную разреженную one-hot матрицу.

### Параметры и градиенты
- таблица должна быть `nn.Parameter`, иначе optimizer не увидит её;
- индексы должны иметь целочисленный dtype, обычно `torch.long`;
- допустимые значения индексов: от `0` до `V-1`;
- после backward градиенты получают строки, которые участвовали в lookup.

### План реализации
1. Создайте матрицу весов формы `(num_embeddings, embedding_dim)`.
2. Зарегистрируйте её как обучаемый параметр модуля.
3. В `forward` используйте тензорную индексацию, сохраняя все ведущие размерности indices.

### Контракт
```python
class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings: int, embedding_dim: int): ...
    def forward(self, indices: Tensor) -> Tensor: ...
```

### Ограничения
- `self.weight` — `nn.Parameter` формы `(num_embeddings, embedding_dim)`;
- forward выполняет lookup по индексам;
- не используйте `nn.Embedding`.

### Быстрые самопроверки
- индекс `[0]` возвращает первую строку weight;
- повторяющиеся индексы возвращают одинаковые векторы;
- вход `(2, 3)` даёт выход `(2, 3, D)`;
- `dict(model.named_parameters())` содержит таблицу весов;
- после backward у weight есть градиент.

### Типичные ошибки
- обычный Tensor вместо `nn.Parameter`;
- float-индексы;
- ручное flatten, после которого теряются batch/sequence dimensions;
- перепутанные `num_embeddings` и `embedding_dim`.

### Официальные материалы и примеры
