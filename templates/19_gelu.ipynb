# 🟢 Easy: GELU Activation

Реализуйте точную **GELU** (Gaussian Error Linear Unit). В отличие от ReLU, GELU не обрывает все отрицательные значения жёстким порогом, а плавно масштабирует вход через вероятность стандартного нормального распределения.

$$\text{GELU}(x) = x \cdot \Phi(x) = x \cdot 0.5 \cdot (1 + \text{erf}(x / \sqrt{2}))$$

Здесь `Φ(x)` — CDF стандартного нормального распределения, а `erf` — error function. В этой задаче используется точная формула, не tanh-аппроксимация.

### Интуиция
- при большом положительном `x` множитель `Φ(x)` близок к `1`, поэтому GELU близка к `x`;
- при большом отрицательном `x` множитель близок к `0`;
- около нуля переход плавный, а `GELU(0) = 0`;
- операция поэлементная: shape входа сохраняется.

### План реализации
1. Представьте `sqrt(2)` как числовую константу без создания несовместимого тензора.
2. Вычислите аргумент `erf` для каждого элемента.
3. Соберите формулу только из дифференцируемых тензорных операций.

### Ограничения
- не используйте `F.gelu`, `nn.GELU` или `torch.nn.functional.gelu`;
- используйте `torch.erf` для точной версии.

### Быстрые самопроверки
- `my_gelu(torch.tensor(0.0))` равна нулю;
- форма произвольного тензора сохраняется;
- результат близок к `F.gelu(x, approximate="none")`;
- backward создаёт конечные градиенты.

### Типичные ошибки
- пропущенный множитель `0.5`;
- `sqrt(x)` вместо `sqrt(2)`;
- использование tanh-аппроксимации при требовании точной формулы;
- вычисление через NumPy, разрывающее autograd.

### Официальные материалы и примеры
