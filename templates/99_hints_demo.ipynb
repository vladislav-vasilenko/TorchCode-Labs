# Демонстрация многоуровневых подсказок

Этот демонстрационный ноутбук проверяет механизм `hints` на inline-словаре и будет удалён после появления задач трека.

```python
from torch_judge.hints import get_hints

TASK = {
    "hints": [
        "Сначала уточните контракт функции и ожидаемые формы тензоров.",
        "Разбейте вычисление на небольшие шаги и проверьте каждый отдельно.",
        "Сопоставьте каждый шаг с подходящей операцией PyTorch.",
    ]
}

assert len(get_hints(TASK)) == 3
print("Three hint levels are available.")
```

<details>
<summary>Tip 1</summary>

Сначала уточните контракт функции и ожидаемые формы тензоров.

</details>

<details>
<summary>Tip 2</summary>

Разбейте вычисление на небольшие шаги и проверьте каждый отдельно.

</details>

<details>
<summary>Tip 3</summary>

Сопоставьте каждый шаг с подходящей операцией PyTorch.

</details>
