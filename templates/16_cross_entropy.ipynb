# 🟢 Easy: Cross-Entropy Loss

Реализуйте **cross-entropy loss** для многоклассовой классификации. На вход подаются logits — ненормированные оценки классов, а не вероятности. Чем выше logit правильного класса относительно остальных, тем меньше loss.

$$\text{CE}(x, y) = -\log\frac{e^{x_y}}{\sum_j e^{x_j}}$$

Эквивалентная и более удобная для реализации форма:

$$\text{CE}(x, y) = \operatorname{logsumexp}(x) - x_y$$

### Формы и смысл данных
- `logits`: `(B, C)`, где `B` — размер batch, `C` — число классов;
- `targets`: `(B,)`, целые индексы правильных классов от `0` до `C-1`;
- результат: один scalar — среднее значение loss по batch.

Не применяйте Softmax заранее: cross-entropy должна получить исходные logits и сама выполнить устойчивую log-нормализацию.

2. Выберите logit правильного класса согласно `targets`.
3. Вычтите выбранный logit из `logsumexp`.
4. Усредните значения по batch.

### Ограничения
- не используйте `F.cross_entropy` или `nn.CrossEntropyLoss`;
- вычисление должно оставаться устойчивым для logits большой величины.

### Быстрые самопроверки
- для одинаковых logits loss равен примерно `log(C)`;
- увеличение logit правильного класса уменьшает loss;
- прибавление одной константы ко всем классам не меняет loss;
- результат совпадает с `F.cross_entropy` после самостоятельной реализации.

### Типичные ошибки
- передача one-hot targets вместо индексов;
- Softmax перед cross-entropy;
- выбор элемента по неверной оси;
- `exp` больших logits без стабилизации.
