# 🟢 Easy: Gradient Norm Clipping

Реализуйте **gradient norm clipping** — ограничение общей нормы градиентов перед шагом optimizer. Это защищает обучение от редких слишком больших обновлений, не меняя направление совокупного градиента.

### Общая норма
Считайте градиенты всех параметров частями одного длинного вектора. Для L2-нормы:

$$\lVert g \rVert_2 = \sqrt{\sum_p \lVert p.\mathrm{grad} \rVert_2^2}$$

Если `||g|| <= max_norm`, градиенты остаются без изменений. Иначе каждый градиент умножается на один коэффициент:

$$\alpha = \frac{\text{max\_norm}}{\lVert g \rVert_2}$$

Одинаковый `α` сохраняет направление вектора, меняя только его длину. Clipping выполняют после `backward()` и до `optimizer.step()`.

### Быстрые самопроверки
- большая норма после вызова не превышает `max_norm`;
- малая норма и сами градиенты остаются неизменными;
- нормализованный градиент до и после clipping совпадает;
- возвращается исходная, а не новая норма.

### Типичные ошибки
- clipping параметров вместо `p.grad`;
- отдельное ограничение каждого тензора вместо одной глобальной нормы;
- вызов до backward, когда gradients ещё нет;
- возврат нормы после масштабирования.
